## Append Table

```python
df = pd.concat([data1, data2, data3, data4])
```

```python
df
```

## Remove the XXX

```python
xxx_index = df[df['Order ID'] == 'XXXXXX'].index
```

```python
df.drop(xxx_index,inplace=True)
```

```python
df
```

## Splitting of Customer id and Customer name

```python
df[['Customer ID', 'Customer Name']] = df['Customer ID'].str.split(' - ' , expand = True)
```

```python
df
```

## Show Total Profit generated from each customer

```python
df['profit'] = df['Revenue'] - df['Cost']
```

```python
df
```

## Total profit for each customer

```python
df.groupby('Customer Name',as_index=False)['profit'].sum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['Customer Name'],df['profit'])
plt.xticks(rotation=45)

plt.show()
```

## average proportion cookies shipped for order

```python
df.groupby('Order Status',as_index=False)['Cookies Shipped'].mean()
```

```python
d = df.groupby('Order Status',as_index=False)['Cookies Shipped'].mean()
```

```python
plt.pie(labels=d['Order Status'],x=d['Cookies Shipped'])
plt.show()
```
